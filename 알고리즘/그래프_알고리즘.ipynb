# 그래프 알고리즘
## 그래프 알고리즘이란?
+ 정점과 이 점들을 연결하는 간선으로 구성된 자료 구조
### 그래프 알고리즘의 중요성
- 최적의 경로 발견
    - 다양한 경로 중 최적(최단거리, 시간, 비용 등)의 경로를 찾아냄
        - ex) GPS 네비게이션, 물류 및 배송 경로 최적화, 전력망, 수도망 등
    - 일상생활에서의 연결망 파악
        - 복잡한 관계망을 분석 및 유용한 정보 추출
    - 인공지능 기술과의 결합
        - 그래프 데이터의 유용한 정보 추출
### 용어 정리
- 그래프 : 정점(node), 간선(edge)로 구성된 자료구조
- 정점 : 그래프의 기본 단위로 위치나 교차점을 나타냄
- 간선 : 각 정점을 연결하는 선
    - 가중치 : 간선의 비용, 시간, 거리 등의 값을 표현
- 무향 그래프 : 방향이 없는 그래프
- 유향 그래프 : 방향이 있는 그래프
### 그래프의 표현
- 인접 행렬
    - 정의 : 그래프의 노드들 간의 연결 관계를 행렬로 나타냄
        - 인접 행렬에서 각 요소는 두 정점의 연결 상태를 나타냄
    - 장점 : 직관적이며 쉽게 구현 가능
    - 단점 : 연결되지 않은 노드들에 대한 정보의 저장 (메모리 과다)
- 인접 리스트
    - 정의 : 각 노드 별로 인접한 모든 노드를 리스트로 나타냄
    - 장점 : 연결된 정보만을 저장하여 메모리 절약
    - 단점 : 인접행렬에 비해 병렬 처리 하기 어려움 (순차적으로 처리해야하고, 동시에 처리 불가)
### 그래프의 탐색
- 하나의 정점에서 시작하여 그래프의 모든 정점을 한 번씩 방문
- 탐색 방법
    - 깊이 우선 탐색 (DFS)
    - 너비 우선 탐색 (BFS)
### 깊이 우선 탐색
- 설명 : 시작 노드에서 한 방향으로 갈 수 있는 가장 멀리 있는 노드에 도달할 때 까지 탐색
        더이상 탐색할 노드가 없으면 마지막으로 방문한 노드로 되돌아와 다른 방향으로 탐색
- 활용 : 경로 존재 확인 (미로찾기), 백트래킹(특정 조건 만족하는 모든 해 찾기)
- 구현 방법 : 재귀 함수, 스택
- 시간 복잡도 : 인접 행렬 O(n^2)
### 요점 정리
- 그래프 : 노드와 간선으로 이루어진 자료구조
- 그래프 표현 방식 : 인접 행렬, 인접 리스트
- 그래프 탐색 방법 : DFS, BFS
