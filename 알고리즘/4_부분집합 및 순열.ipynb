```python
# 부분집합을 재귀함수를 통해 구하기
# 아이디어: 
# 각 리스트의 요소마다 0 또는 1의 bit값을 생성해서 부분집합을 표현
# 재귀함수를 통해서 깊이(인덱스)를 순차적으로 주어서 bit 값을 표현

def print_subset(bits, depth):
    for i in range(2): # 0 또는 1 비트값 생성
        bits[depth] = i     #현재 비트값 할당
        
        if depth == len(bits)-1:     #깊이가 끝까지 도달한 경우 출력
            print(bits)
        else:
            print_subset(bits, depth+1)
bits = [0, 0, 0]
print_subset(bits)
```

```python
# 부분집합 중 타겟의 합과 같은 부분집합만 출력
# 가지치기 : 
# 이전 부분집합의 합이 target보다 큰 경우 탐색 더 안 함
# 이전 부분집합의 합 + 남아 있는 모든 요소의 합이 타겟보다 작은 경우 탐색 더 안 함

def print_subset(bits, depth, pre_sum, target) :
    for i in range(2) : # 0 또는 1 비트값 생성
        bits[depth] = i # 현재 비트값 할당
        cur_sum = pre_sum + A[depth]*bits[depth]
        
        # 타겟과 현재 합이 일치하는 경우
        if cur_sum == target:
            for j in range(depth+1): # 부분집합 출력
                temp = []
                if bits[j]:
                    print(A[j], end = ' ')
                    temp.append(A[j])
            print()
            subset.add(tuple(temp))
            
        # 일치하지 않은 경우 더 탐색
        # cur_sum<=target : 부분집합에서 0이 여러 개 있을 수 있는 경우가 있기 때문에 등호를 꼭 넣어주어야 한다
        # ex) A = [20, 0]  ##(20), (20, 0)
        # 아래 코드가 없어도 실행되지만 가지치기로 실행속도 줄이기 위해 쓰는 코드
        if cur_sum <= target and cur_sum + sum(A[depth+1:]) >= target :  
              if depth < len(bits)-1: 
                print_subset(bits, depth+1, cur_sum, target)     
        
subset = set()       
A = [i for i in range(10)]
bits = [0]*len(A)
print_subset(bits, 0, 0, 1)
print(subset)
```

```python
# 순열 표현
# 아이디어:
# 순열의 각 자리마다 인덱스 가짓수 만큼 경우의 수가 있음(=순열 크기 만큼 있음)
# 이전 순열요소와 중복되지 않아야 함
# 중복되 경우는 탐색x, 중복이 안된 경우만 탐색

# 순열의 인덱스를 출력
def print_permutation(idxs, depth):
    # 연재 깊이(인덱스) 순열 현재 인덱스 결정
    for i in range(len(idxs)):
        # 이전 번호와 중복되는지 체크
        duplicate = False
        for d in range(depth):
            if idxs[d] == i:
                break
        if duplicate : continue
        
        idxs[depth] = i     #현재 깊이의 인덱스 값 할당
        if depth == len(idxs)-1:
            print(idxs)
            for j in idxs:   #인덱스 순서대로 출력
                print(fruits[j], end=' ')
            print()
        else:
            print_permutation(idxs, depth+1)
            
fruits = ['사과', '딸기', '포도']
idxs = [0]*len(fruits)
print_permutation(idxs, 0)
        
```
